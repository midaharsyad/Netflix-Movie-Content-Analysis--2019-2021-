#Data Cleaning

##Filter Data Movie

```python
df_movie = df_netflix[df_netflix['type'] == 'Movie']
df_movie.head()
```

```python
df_movie.info()
```

##Dataset Movie Tahun 2019-2021

```python
df_movie = df_movie[df_movie['years_added'].between(2019,2021)]
df_movie.head()
```

```python
df_movie.info()
```

##Menghapus kolom yang tidak digunakan dalam analisis

```python
df_movie.drop(columns = ['title','cast','date_added','release_year','description'], inplace = True)
```

```python
df_movie.columns
```

##Missing Value

```python
df_movie.info()
```

```python
df_movie.isnull().sum()
```

```python
df_movie = df_movie[~df_movie['country'].isna()] #Exclude all null data
df_movie.head()
```

```python
df_movie.isnull().sum()
```

```python
df_movie = df_movie[~df_movie['director'].isna()] #Exclude all null data
df_movie.head()
```

```python
df_movie.isnull().sum()
```

##Duplicate Data

```python
df_movie[df_movie['show_id'].duplicated()]
```

```python
df_movie[df_movie.columns[3]].value_counts()
```

```python
df_movie[df_movie.columns[7]].value_counts()
```

```python
df_moviecountry = df_movie['country'].str.split(",",expand = True)
df_moviecountry.head()
```

```python
df_moviecountry = df_movie['country'].str.split(",",expand = True).rename(columns={0:'country1',1:'country2',2:'country3',3:'country4',4:'country5',5:'country6',6:'country7',7:'country8'})
df_moviecountry.head()
```

```python
df_moviegenre = df_movie['listed_in'].str.split(",",expand = True)
df_moviegenre.head()
```

```python
df_moviegenre = df_movie['listed_in'].str.split(",",expand = True).rename(columns={0:'genre1',1:'genre2',2:'genre3'})
df_moviegenre.head()
```

```python
df_movieclean = pd.concat([df_movie,df_moviecountry,df_moviegenre],axis=1)
df_movieclean.head()
```

```python
df_movieclean.drop(columns = ['country','listed_in'], inplace = True)
```

```python
df_movieclean.columns
```

```python
df_movieclean.head()
```

#EDA & Visualisasi Data

##Total Movie per Tahun

```python
df_totalmovie = df_movieclean['years_added'].value_counts()
df_totalmovie
```

```python
colors = ['#FFC5C5', '#FFEBD8', '#C7DCA7', '#89B9AD']

jumlahmovie = df_totalmovie.head(5)

plt.figure(figsize=(10, 5))
sns.barplot(x=jumlahmovie.index, y=jumlahmovie.values, palette=colors)
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.title('Jumlah Movie per Tahun')
plt.show()
```

##Total Movie by country

```python
df_country = df_movie['country'].str.split(', ').explode().value_counts().reset_index(name='Jumlah')
df_country.head(10)
```

```python
topcountry = pd.DataFrame(df_country).rename(columns={'index':'Country'})
topcountry
```

```python
colors = ['#FFC5C5', '#FFEBD8', '#C7DCA7', '#89B9AD']
top_country = topcountry.head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_country.country, y=top_country.Jumlah, palette=colors)
plt.xlabel('Country')
plt.ylabel('Jumlah')
plt.title('Top 10 Country')
plt.show()
```

```python
df_movieclean['All_Countries'] = df_movieclean[['country1', 'country2', 'country3','country4','country5', 'country6', 'country7', 'country8']].apply(lambda x: ', '.join(x.dropna()), axis=1)
df_movieclean['All_Genres'] = df_movieclean[['genre1', 'genre2', 'genre3']].apply(lambda x: ', '.join(x.dropna()), axis=1)
df_melted = pd.melt(df_movieclean, id_vars=['show_id', 'All_Genres'], value_vars=['country1', 'country2', 'country3','country4','country5', 'country6', 'country7', 'country8'], var_name='Country', value_name='Country_Value')
df_us = df_melted[(df_melted['Country_Value'].notna()) & (df_melted['Country_Value'] == 'United States')]
df_us['All_Genres'] = df_us['All_Genres'].str.split(', ')
df_us_exploded = df_us.explode('All_Genres')
jumlah_genre_us = df_us_exploded['All_Genres'].value_counts()
jumlah_genre_us
```

```python
genreus = pd.DataFrame(jumlah_genre_us)
genreus
```

```python
genreus = jumlah_genre_us.reset_index()
genreus.columns = ['Genre', 'Jumlah']

colors = ['#FFC5C5', '#FFEBD8', '#C7DCA7', '#89B9AD']

top_genreus = genreus.head(5)

plt.figure(figsize=(10,3))
sns.barplot(x='Jumlah', y='Genre', data=top_genreus, palette=colors)

plt.xlabel('Jumlah')
plt.ylabel('Genre')
plt.title('Top 10 Genre di United States')
plt.show()
```

##Total Movie by Genre

```python
df_genre = df_movie['listed_in'].str.split(', ').explode().value_counts().reset_index(name='Jumlah')
df_genre
```

```python
topgenre = df_genre.rename(columns={'listed_in':'Genre'})
topgenre
```

```python
colors = ['#FFC5C5', '#FFEBD8', '#C7DCA7', '#89B9AD']

top_genre = topgenre.head(10)

plt.figure(figsize=(10,5))
sns.barplot(x='Jumlah', y='Genre', data=top_genre, palette=colors)

plt.xlabel('Jumlah')
plt.ylabel('Genre')
plt.title('Top 10 Genre')
plt.show()
```

```python
df_movie = df_movie.explode('country')
df_movie['listed_in'] = df_movie['listed_in'].apply(lambda x: str(x))
df_genres = df_movie['listed_in'].str.split(', ', expand=True)
df_genres['country'] = df_movie['country']
df_expanded = df_genres.melt(id_vars='country', value_name='listed_in').dropna()
genre_1 = 'Dramas'
genre_2 = 'International Movies'
genre_3 = 'Comedies'
df_selected_genres = df_expanded[df_expanded['listed_in'].isin([genre_1, genre_2,genre_3])]
jumlah_film_negara_genre = df_selected_genres.groupby(['country']).size().reset_index(name='Jumlah_Film')
jumlah_film_negara_genre
```
